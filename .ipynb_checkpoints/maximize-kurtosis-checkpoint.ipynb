$$
p(s_i) = 
  \begin{cases}
    \frac{1}{2 \sqrt{3}}, (|s_i| \leq \sqrt{3}) \\
    0, (other)
  \end{cases}
$$

$$
\boldsymbol{S} = \left(
    \begin{array}{cc}
      s_1 \\
      s_2
    \end{array}
  \right)
$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def theta2matrix(theta):
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)],
    ])

def theta2xy(theta):
    return np.cos(theta), np.sin(theta)

def get_kurtosis(theta, Z):
    W = theta2matrix(theta)
    # 回転行列から独立変数を取り出す
    w = W[:, 0]
    y = np.dot(w.T, Z)
    return kurtosis(y)

def plot(Z):
    plt.figure(figsize=(5, 5))
    plt.scatter(Z[0, :], Z[1, :], s=1)
    disp_xy = np.max(Z) * 2
    plt.xlim(-disp_xy, disp_xy)
    plt.ylim(-disp_xy, disp_xy)
    plt.show()

# 一様分布にはuniformを使う
I = 2
# サンプル数
N = 1000
S = np.random.uniform(-np.sqrt(3), np.sqrt(3), [I, N])
s_1 = S[0, :]
s_2 = S[1, :]
display("mean: %.3f, var: %.3f" % (np.mean(s_1), np.var(s_1)))
display("mean: %.3f, var: %.3f" % (np.mean(s_2), np.var(s_2)))

plot(S)
```

$$\boldsymbol{A} = \left(
    \begin{array}{cc}
      5 & 10 \\
      10 & 2
    \end{array}
  \right) $$
  
$$
\boldsymbol{X} = \boldsymbol{AS}
$$

```python
A = np.array([
    [5, 10],
    [10, 2]
])
A = np.dot(
    theta2matrix(0.5),
    A)

X = np.dot(A, S)
plot(X)
```

$$\Sigma = E\{x x^T\}$$

```python
Sigma = np.dot(X, X.T) / N
Sigma
```

* $E$は$\Sigma = E\{x x^T\}$の固有ベクトルからなる直交行列
* $D$は$\Sigma = E\{x x^T\}$の固有値からなる対角行列
* $D ^ {-\frac{1}{2}} = diag(d_1^{-\frac{1}{2}}, ..., d_n^{-\frac{1}{2}})$

```python
from IPython.display import Math

# 固有値計算
d, E = np.linalg.eig(Sigma)
D = np.diag(d)
D_sqinv = np.diag(1 / np.sqrt(d))
# 同じ
# D_sqinv = np.linalg.inv(np.sqrt(D))
display(Math("D"), D)
display(Math("D^{-\\frac{1}{2}}"), D_sqinv)

display(Math("\\Sigma"), Sigma)
display(Math("DED^T = \\Sigma"), np.dot(E, np.dot(D, E.T)))

V = np.dot(
    E,
    np.dot(
        D_sqinv, 
        E.T))
display(Math("V"), V)
```

$$\boldsymbol{Z} = \boldsymbol{V}\boldsymbol{X}$$

```python
Z = np.dot(V, X)

plot(Z)
```

$kurt(w^Tz) = E\{(w^Tz)^4\} -3||w||^4$

$\frac{\partial}{\partial w} |kurt(w^T z)|$

## $||w||^4$について

$\frac{d}{dx}f(g(x)) = f'(g(x))g'(x)$

$g'(x) = \frac{\partial}{\partial w} ||w||^2  = \frac{\partial}{\partial w} w^T w = 2 w$

$\frac{\partial}{\partial w} ||w||^4  = \frac{\partial}{\partial w} (||w||^2)^2 = 4w||w||^2$

## signについて

sgnとも。絶対値の働きをする

$$
sign(x) = \begin{cases}
1 \  (x > 0) \\
0 \  (x = 0) \\
-1 \  (x < 0)
\end{cases}
$$

```python
from tqdm import tqdm

L = 100
alpha = 0.1
w = np.array([1.0, 0.0])

l = []

for n in range(L):
    # y_hatは独立成分のサンプル...確率変数
    y_hat = np.dot(w.T, Z)
    l.append(kurtosis(y_hat))
    # kurtosisでサンプルの尖度を測る
    # dotだと合計されるので、shape[-1]で除算し平均を取る
    dw = np.sign(kurtosis(y_hat)) * (np.dot(Z, np.power(y_hat, 3)) / y_hat.shape[-1])
    w = w + alpha * dw
    w = w / np.linalg.norm(w)

plot(Z)

plt.plot(l)
plt.show()

# 変換前と変換後比較
plt.hist(Z[0, :])
plt.show()
plt.hist(y_hat)
plt.show()
```

```python
# すべての成分について一挙に解く

L = 100
alpha = 0.1
W = np.array([
    [1.0, 0.0],
    [0.0, 1.0]
 ])
W = W / np.linalg.norm(W, axis=1)

l = []

for n in range(L):
    Y_hat = np.dot(W.T, Z)
    dW = np.dot(
        np.sign(kurtosis(Y_hat, axis=1)),
        np.dot(Z, np.power(Y_hat, 3).T) / Y_hat.shape[-1]
    )
    W = W + alpha * dW
    W = W / np.linalg.norm(W, axis=1)
    l.append(np.sum(kurtosis(Y_hat, axis=1)))

plot(Z)
Y_hat = np.dot(W.T, Z)
plot(Y_hat)

plt.plot(l)
plt.show()
```
